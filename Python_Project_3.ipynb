```python
import io
df = pd.read_csv(io.BytesIO(uploaded["NYPD_Arrest_Data (1).csv"]))
df.head()
```

```python
df.groupby(["Boro"])["Boro"].count()
```

```python
#Seperating the Boroughs at multiple df's
dfm = df[(df["Boro"])== "M"]
dfb = df[(df["Boro"])== "B"]
dfk = df[(df["Boro"])== "K"]
dfq = df[(df["Boro"])== "Q"]
dfs = df[(df["Boro"])== "S"]
```

```python
#Creating a figure to see what age people in Brooklyn are most likely to commit crimes
mp.figure(figsize=(10,6))
mp.hist((dfk["Age"]), bins = 12, alpha = 0.6, density = False)
mp.show()
dfk["Age"].describe()
```

```python
#Finding the age at which people in Manhattan are more likely to commit crimes
mp.figure(figsize=(10,6))
mp.hist((dfm["Age"]), bins = 12, alpha = 0.6, density = False)
mp.show()
dfm["Age"].describe()
```
